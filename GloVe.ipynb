## Le constat initial (tableau)

Le tableau donne, pour différents mots *k* (solid, gas, water, fashion), les probabilités conditionnelles *P(k|ice)* et *P(k|steam)*, ainsi que leur rapport.

| k | P(k\|ice) | P(k\|steam) | ratio = P(k\|ice) / P(k\|steam) |
|---|-----------|-------------|-------|
| solid | 1.9×10⁻⁴ | 2.2×10⁻⁵ | 8.9 |
| gas   | 6.6×10⁻⁵ | 7.8×10⁻⁴ | 0.085 |
| water | 3.0×10⁻³ | 2.2×10⁻³ | 1.36 |
| fashion | 1.7×10⁻⁵ | 1.8×10⁻⁵ | 0.96 |

- **solid** est lié à *ice* (on dit “ice solid”), pas à *steam* → ratio grand (>1).
- **gas** est lié à *steam* (“steam gas”), pas à *ice* → ratio petit (<1).
- **water** est lié aux deux → ratio ≈ 1.
- **fashion** n’est lié ni à l’un ni à l’autre → ratio ≈ 1.

Ce qui compte, ce n’est pas la valeur absolue des probabilités, mais **le rapport** *P(k|ice) / P(k|steam)*. Ce rapport est ce qui permet de distinguer les mots pertinents pour *ice* de ceux pertinents pour *steam*.

## Le lien avec la factorisation

GloVe veut apprendre des plongements $w$ (pour les mots) et $\tilde{w}$ (pour les contextes) de sorte que le produit scalaire $w_i · \tilde{w}_k$ soit proche de $log(X_ik)$, où $X_ik$ est le nombre de fois que $k$ apparaît dans le contexte de $i$.

Or, $P(k|i) = X_ik / X_i$ (où $X_i$ est le nombre total de contextes du mot $i$). Donc :
$$
\frac{P(k|i)}{P(k|j)} = \frac{X_ik / X_i}{X_jk / X_j} = \frac{X_ik}{X_jk} \cdot \frac{X_j}{X_i}.
$$

En prenant le logarithme :
$$
\log\frac{P(k|i)}{P(k|j)} = \log X_ik - \log X_jk + \log X_j - \log X_i.
$$

Si on suppose que le modèle apprend $w_i · \tilde{w}_k ≈ \log X_ik$ (à un biais près), alors la différence
$$
(w_i · \tilde{w}_k) - (w_j · \tilde{w}_k) \approx \log X_ik - \log X_jk.
$$

Avec les biais $b_i$ et $\tilde{b}_k$ qui absorbent les termes $\log X_i$ et $\log X_j$, on obtient que **la différence entre les produits scalaires encode directement le log du rapport de probabilités** :
$$
(w_i - w_j) · {w}_k \approx \log\frac{P(k|i)}{P(k|j)}.
$$

Ainsi, lorsque $k$ est fortement associé à $i$ mais pas à $j$, le rapport est grand → le produit scalaire $(w_i - w_j)·\tilde{w}_k$ est grand et positif. Si $k$ est associé à $j$, le rapport est faible → le produit scalaire est négatif. Si $k$ est associé aux deux ou à aucun, le rapport est proche de 1 → le produit scalaire est proche de 0.

## En résumé

La factorisation de GloVe n’est pas un simple calcul numérique : elle est **construite pour reproduire ces rapports** dans l’espace vectoriel. C’est cette propriété qui donne aux plongements leur pouvoir de capturer des relations sémantiques fines (analogies, oppositions, etc.). Le tableau est donc la preuve empirique que les rapports de cooccurrence contiennent l’information sémantique, et l’objectif de GloVe est de projeter cette information dans un espace dense de petite dimension.
