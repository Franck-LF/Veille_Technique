```python
import gensim

import gensim.downloader as api
from multiprocessing import cpu_count
from gensim.models.word2vec import Word2Vec

dataset = api.load("text8")
data =[]
for word in dataset:
  data.append(word)

data_1 = data[:1200] 
data_2 = data[1200:] 

# Training the Word2Vec model
w2v_model = Word2Vec(data_1, min_count = 0, workers = cpu_count())
print(w2v_model['time'])
```

```python
from gensim.models import Word2Vec

corpus = [["J'", "aime", "ce", "film"],
          ["Ce", "film", "est", "excellent"],
          ["Le", "résumé", "est", "confus"]]

# Modèle Skip-gram (sg = 1)
model = Word2Vec(sentences=corpus, vector_size = 200, min_count=1, sg=0)
print('Size:', len(model.wv.get_vector('aime')))

for word in model.wv.key_to_index:
    print(word, model.wv.get_vector(word))
```
