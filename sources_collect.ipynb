```python
import arxiv

# Construire une requête
search = arxiv.Search(
  query = "machine learning",
  max_results = 5,
  sort_by = arxiv.SortCriterion.SubmittedDate
)

# Itérer sur les résultats
for result in search.results():
  print(result.title)
```

```python
import arxiv
search = arxiv.Search(
    query="vectorization",
    max_results=15,
    sort_by=arxiv.SortCriterion.SubmittedDate
)

for paper in search.results():
    print(f"Titre : {paper.title}")
    print(f"Auteurs : {paper.authors}")
    print(f"Résumé : {paper.summary[:200]}...")
    print(f"Lien : {paper.pdf_url}\n")
```
