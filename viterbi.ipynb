# Exemple Viterbi

```python
# -----------------------------
# Exemple Viterbi en Python
# -----------------------------

# États cachés
states = ["Soleil", "Pluie"]

# Observations possibles
observations = ["Lunettes", "Parapluie"]

# Séquence observée
sequence_obs = ["Lunettes", "Parapluie", "Parapluie"]
# sequence_obs = ["Lunettes", "Parapluie", "Parapluie", "Parapluie", "Parapluie"]
sequence_obs = ["Lunettes", "Parapluie", "Lunettes", "Parapluie"]

# Probabilités initiales (P(X0))
start_prob = {
    "Soleil": 0.6,
    "Pluie": 0.4
}

# Probabilités de transition (A)
trans_prob = {
    "Soleil": {"Soleil": 0.7, "Pluie": 0.3},
    "Pluie": {"Soleil": 0.4, "Pluie": 0.6}
}

# Probabilités d'émission (B)
emit_prob = {
    "Soleil": {"Lunettes": 0.9, "Parapluie": 0.1},
    "Pluie": {"Lunettes": 0.2, "Parapluie": 0.8}
}

# -----------------------------
# Algorithme de Viterbi
# -----------------------------

V = [{}]  # tableau des probabilités maximales
chemins = {}  # pour backtracking

# Initialisation
for state in states:
    V[0][state] = start_prob[state] * emit_prob[state][sequence_obs[0]]
    chemins[state] = [state]

# Récursion
for t in range(1, len(sequence_obs)):
    V.append({})
    new_chemins = {}
    
    for curr_state in states:
        # On choisit le meilleur état précédent
        (prob, prev_state) = max(
            (V[t-1][prev_state] * trans_prob[prev_state][curr_state] * emit_prob[curr_state][sequence_obs[t]], prev_state)
            for prev_state in states
        )
        V[t][curr_state] = prob
        new_chemins[curr_state] = chemins[prev_state] + [curr_state]
    
    chemins = new_chemins

# Terminaison
(n, dernier_etat) = max((V[-1][state], state) for state in states)

print("Séquences observées :\n", sequence_obs)
print("Meilleure séquence d'états cachés :\n", chemins[dernier_etat])
print("Probabilité de cette séquence :", n)
```
